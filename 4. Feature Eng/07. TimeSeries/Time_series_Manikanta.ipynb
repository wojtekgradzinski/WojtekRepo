```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df1= pd.read_csv(r'climate.csv')
df1
```

```python
df1=df1.drop(['Date Time'], axis=1)
df1
```

```python


ss=StandardScaler()
df=pd.DataFrame(ss.fit_transform(df1), columns= df1.columns)
df
```

```python
df.shape
```

```python
def pairing(data, seq_length=6):
    x=[]
    y=[]
    for i in range(0, (df.shape[0]-1)//seq_length, seq_length+1):
        seq= np.zeros((seq_length, df.shape[1]))
        for j in range(seq_length):
            seq[j]= data.values[i+j]
        x.append(seq)
        y.append(df['T (degC)'][i+seq_length])
    return np.array(x), np.array(y)
x,y = pairing(df)
print(x.shape)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
nsamples, nx, ny = X_train.shape
X_train_d2 = X_train.reshape((nsamples,nx*ny))
X_train_d2
```

```python
nsamples, nx, ny = X_test.shape
X_test_d2 = X_test.reshape((nsamples,nx*ny))
X_test_d2.shape
```

```python
 np.shape(X_train_d2)
```

```python
# y_train=np.reshape(y_train, (-1, 2))

y_train=y_train.reshape(-1,1)
y_train.shape
y_test=y_test.reshape(-1,1)
```

```python
from sklearn.ensemble import RandomForestRegressor

clf= RandomForestRegressor()
clf.fit(X_train_d2, y_train)

```

```python
from sklearn.metrics import mean_absolute_error
pred= clf.predict(X_test_d2)
mean_absolute_error(y_test, pred)
```
