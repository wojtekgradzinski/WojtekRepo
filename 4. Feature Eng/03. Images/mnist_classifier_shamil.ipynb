## 2.- As you can see each class has its own folder (Do it only for train). 

    - Iterate folder by folder ( os.listdir() )
    - Inside each folder: 
        1.- Read the image
        2.- Reshape it into a flat array (784,)
        3.- Save the data into a pandas dataframe apending the column name as the class
    - Save the data into a CSV

    Note: if it takes to long try doing only 100 images per folder and the teacher for the CSV.

```python
parent_dir = 'Data'
df = pd.DataFrame()
for folder in os.listdir('Data'):
    if folder == 'trainingSet':
        for f in os.listdir('Data'+'/'+folder+'/'+folder):
            if f == '0':
                # print(type(len('Data'+'/'+folder+'/'+folder+'/'+'0')))
                class_data = np.zeros((len(os.listdir('Data'+'/'+folder+'/'+folder+'/'+f)),785)) #rows is the length of the list of elements on a folder

                for i,image in enumerate(os.listdir('Data'+'/'+folder+'/'+folder+'/'+f)):
                    img = Image.open('Data'+'/'+folder+'/'+folder+'/'+f+'/'+image)
                    print(img.size)
                    img_resized = np.array(img,dtype = 'int')
                    # print(img_resized)
                    img_resized = img_resized.flatten()
                    print(np.array(img_resized.size))
                    # print(img_resized)
                    class_data[i,:784] = img_resized
                    class_data[i,784] = int(f)
                class_data = pd.DataFrame (class_data)
                df = pd.concat ([df, class_data])
df.to_csv('data.csv',index=False)
# print()
```
