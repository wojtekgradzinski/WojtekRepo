```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def imshow(img, grayscale = False):
    plt.figure(figsize = (15,10));
    if grayscale:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img[:,:,::-1])    
```

```python
# With numpy create a black canvas of size 800 x 800 
canvas = np.zeros((800,800), dtype=np.uint8)

canvas = cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)

imshow(canvas)
```

```python


# Make a copy of the canvas and draw a white square of size 250 x 250 on top of the copy of the canvas

rectangle = canvas.copy()
color= (255,255,255)
square = cv2.rectangle(rectangle,(0,0),(250,250),color,cv2.FILLED)

imshow(square)
```

```python
# Make a copy of the origina lcanvas and draw a white circle of size 200 x 200 on top of the copy of the canvas
# Make sure the circle is in a different position of the canvas than the square

circle = canvas.copy()
color= (255,255,255)
cv2.circle(circle,(400,400),150, color, cv2.FILLED)

imshow(circle)
```
