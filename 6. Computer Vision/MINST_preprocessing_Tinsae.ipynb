```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import joblib
loaded_model = joblib.load('minst_XGboost.sav')
```

### Imporing and displaying the image

```python
image = cv2.imread('img/raw numbers/9.jpg',0)
plt.imshow(image,cmap='gray');
```

### Crop the image in square & putting the number in the middle

```python
image_croped = image[500:1100,300:800]
plt.imshow(image_croped,cmap='gray');
```

```python
thresh = cv2.adaptiveThreshold(image_croped, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY_INV, 199, 21)
plt.imshow(thresh,cmap='gray');                                       
```

```python
kernel = np.ones((2,2),np.uint8)
dilated = cv2.dilate(thresh,kernel,iterations=50)
plt.imshow(dilated,cmap='gray')
```

```python
resized = cv2.resize(dilated, (28,28), interpolation = cv2.INTER_AREA)
plt.imshow(resized,cmap='gray')
```

```python
cv2.imwrite('img/preprocessed/9.jpg',resized)
```

```python
image_processed = cv2.imread('img/preprocessed/9.jpg',0)
img_arr = image_processed.flatten()
```

```python
kernel = np.ones((2,2),np.uint8)
dilated = cv2.dilate(image_processed,kernel,iterations=0)
plt.imshow(dilated,cmap='gray')
```
