## Erosion

```python
# Using the same kernels, apply Erosion to the thresholded image, make sure to start with a clean image
img = cv2.imread('img/scan.jpg',0)

th = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV, 23,5)

#ret, th = cv2.threshold(img,220,255,cv2.THRESH_BINARY_INV)
kernel= cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
#erosion = cv2.erode(th,kernel)
dilate = cv2.dilate(th, kernel, iterations=1)

img_new = cv2.bitwise_not(dilate)



cv2.imshow('img',img_new)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Morphological Transformations: Opening / Closing

```python
# Apply opening to the thresholded image, experiment with different iterations 


kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
closing = cv2.morphologyEx(th, cv2.MORPH_CLOSE, kernel, iterations=1)


cv2.imshow('img',closing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
