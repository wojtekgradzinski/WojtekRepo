```python
"""SEABORN PLOTTINGS"""

import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(data= titanic, x ="sex",hue= 'pclass')
sns.set(font_scale=1,palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data= titanic, y ="sex")
plt.show()
```

```python
"""CATEGORICAL PLOTS"""

plt.figure(figsize=(12,6))
sns.stripplot(data= titanic, x ="sex",y = 'age', hue= 'pclass', dodge= True)

plt.style.use('seaborn-deep')
plt.show()
```
