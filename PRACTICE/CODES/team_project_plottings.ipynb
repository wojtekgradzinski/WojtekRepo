```python
import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(data= df, x ="pub_ratings",hue= 'price_range')
sns.set(font_scale=1,palette='viridis')
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.jointplot(data = df, x ='pub_ratings', y='num_reviews', height =8 , kind ='reg')
plt.show()
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data = df, x ='pub_ratings', y='num_reviews', height =8 , col ='price_range')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(data= df, x ="pub_ratings",y = 'num_reviews', hue= 'price_rangerue)

plt.style.use('seaborn-deep')
plt.show();
```
