```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

Total_market_cap = 1656726463
Btc_market_cap =  726000000
Eth_Market_cap =  281285000


btc = Btc_market_cap / Total_market_cap
eth = Eth_Market_cap / Total_market_cap
alts =  1 - ( btc + eth) 

crypto = np.array([btc,eth,alts]).round(3)*100 
labels = ["BTC", "ETH", "ALTS"]
myexplode = [0.1,0,0]

plt.pie(crypto, labels=labels, explode = myexplode, autopct='%1.1f%%', shadow=True)
plt.show()


df = pd.DataFrame(crypto, index=  ["BTC", "ETH", "ALTS"])
df[0] = df[0].apply(lambda x: str(x) + "%")
print(df)
```
