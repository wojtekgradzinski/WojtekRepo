```python
# I know it is not perfect but it is what I was able to do at this level of programming:)

def display_contents_file(file_name):
    fp = open(file_name, "r")
    counts = dict()
    #with open( file_name ) as fp:
    words = fp.read().split()
    words.sort()
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1     
    return counts


    print(words.lower())
    fp.close()
display_contents_file("blakepoems.txt")
```
